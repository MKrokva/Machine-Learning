```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
tables = pd.read_html(url, header=0)

df = tables[2]
```

```python
# 1. Виведіть перші 5 рядків таблиці.
print(df.head(5))
```

```python
# 2. Визначте кількість рядків і стовпчиків за допомогою атрибуту shape.
num_rows, num_columns = df.shape
print("\nКількість рядків:", num_rows)
print("Кількість стовпчиків:", num_columns)
```

```python
# 3. Визначте загальну кількість років та кількість місяців у даних.
total_years = num_rows - 1
total_months = total_years * 12 
print("\nЗагальна кількість років:", total_years)
print("Кількість місяців:", total_months)
```

```python
# 4. Замінити пропущені значення "-" в таблиці на NaN
df.replace("—", np.nan, inplace=True)
df
```

```python
# 5. Визначте тип стовпців за допомогою dtypes.
df.dtypes
```

```python
# 6. Видаліть передостанній і останній стовпчики
df = df.iloc[:, :-2]
df
```

```python
df["В національній валюті"] = df["В національній валюті"].str.replace('▲', "").str.replace('▼', "")
df["В доларах США"] = df["В доларах США"].str.replace('▲', "").str.replace('▼', "")
```

```python
df.dtypes
```

```python
# 7. Замініть тип нечислових даних на числові.
df['В національній валюті'] = df['В національній валюті'].str.replace(',', '.').astype(float)
df['В доларах США'] = df['В доларах США'].str.replace(',', '.').astype(float)
df
```

```python
# 8. Порахуйте частку пропусків в кожній колонці
missing_percentage = df.isnull().sum() / len(df)
print("\nЧастка пропусків в кожній колонці:")
print(missing_percentage)
```

```python
# 9. Замініть всі пропущені значення на середні значення по стовпцю.
df2 = df.fillna(df.mean(), inplace=True)
```

```python
# 10. Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату.
max_salary_year = df.iloc[df['В національній валюті'].idxmax()]['Рік']
max_salary_gn = df['В національній валюті'].max()
print(f"\nРік з найвищою зарплатою в Україні в грн: {max_salary_year}, Зарплата: {max_salary_gn} грн")
```

```python
# 11. Знайдіть рік з мінімальною зарплатою в Україні в дол. США і вкажіть цю зарплату.
min_salary_usd_year = df[df['В доларах США'] == df['В доларах США'].min()]['Рік'].values[0]
min_salary_usd = df['В доларах США'].min()

print(f"\nРік з найменшою зарплатою в Україні дол.США: {min_salary_usd_year}, Зарплата: {min_salary_usd} дол.США")
```

```python
# 12. Виведіть роки, коли зарплата в грн була вище середньої.
above_average_salary_years = df[df["В національній валюті"] > df["В національній валюті"].mean()]["Рік"]
print("\nЗарплата вище середньої:")
print(above_average_salary_years.tolist())
```

```python
# 13. Порівняйте зарплату в грн за останні два доступних роки.
last_two_years_salary = df.iloc[-2:]["В національній валюті"]
print("\nЗарплата в грн за останні два роки:")
print(last_two_years_salary)
```

```python
# 14. Створіть графік, що відображає зарплату в Україні з часом (за роками).
plt.figure(figsize=(20, 5))
plt.plot(df["Рік"], df["В національній валюті"], marker='o')
plt.title("Середня зарплата в Україні за роками")
plt.grid(True)
```

```python
# 15. Видаліть останній рік з таблиці та виведіть оновлену таблицю.
df = df.iloc[:-1]
print("\nОновлена таблиця без останнього року:")
print(df)
```

```python
# 16. Побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року.
plt.figure(figsize=(20, 5))
plt.plot(df["Рік"], df["В національній валюті"], marker='o')
plt.title("Середня зарплата в Україні за роками (без останнього року)")
plt.grid(True)
```

```python
# 17. Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років.
last_two_years_salary = df.iloc[-2:]["В національній валюті"].sum()
last_five_years_salary = df.iloc[-5:]["В національній валюті"].sum()

if last_two_years_salary > last_five_years_salary:
    print("\nЗарплата за останні два роки більше")
else:
    print("\nЗарплата за останні два роки менше")
```
