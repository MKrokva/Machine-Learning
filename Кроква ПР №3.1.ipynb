8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'):

```python
data['video views'].plot(kind='hist', bins=50, edgecolor='k', alpha=0.7)
plt.title('Розподіл переглядів')
plt.show()
```

9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)

```python
max_views = data['video views'].max()
min_views = data['video views'].min()
mean_views = data['video views'].mean()

print('Максимальна кількість переглядів:', max_views)
print('Мінімальна кількість переглядів:', min_views)
print('Середня кількість переглядів:', mean_views)
```

10.Знайдіть країну, де найбільше відео було завантажено на YouTube

```python
most_uploaded_country = data['Country'].value_counts().idxmax()
print('Країна з найбільшою кількістю відео на YouTube:', most_uploaded_country)
```

11. Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads.

```python

max_uploads_index = data['uploads'].idxmax()
min_uploads_index = data['uploads'].idxmin()

video_with_max_uploads = data.loc[max_uploads_index]['Title']
video_with_min_uploads = data.loc[min_uploads_index]['Title']

print('Відео з найбільшою кількістю завантажень:', video_with_max_uploads)
print('Відео з найменшою кількістю завантажень:', video_with_min_uploads)
```
